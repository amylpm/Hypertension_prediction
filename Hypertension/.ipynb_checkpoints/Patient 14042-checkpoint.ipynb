```python
model =  IsolationForest(contamination=0.005)
model.fit(bp[['pulse']])

```

```python
bp['pulse_outliers']=pd.Series(model.predict(bp[['pulse']])).apply(lambda x: 'yes' if (x == -1) else 'no' )
bp.query('pulse_outliers=="yes"')
```

```python
fig = px.scatter(bp.reset_index(), x='recordTime', y='pulse', color='pulse_outliers', title='')
fig.update_xaxes(
    rangeslider_visible=True,
)
fig.show()
```

```python
data = bp[['pulse','systolic','diastolic']]
n_cluster = range(1, 20)
kmeans = [KMeans(n_clusters=i).fit(data) for i in n_cluster]
scores = [kmeans[i].score(data) for i in range(len(kmeans))]
fig, ax = plt.subplots(figsize=(10,6))
ax.plot(n_cluster, scores)
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show();
```

```python
segment_len = 32
slide_len = 2

segments = []
for start_pos in range(0, len(bp), slide_len):
    end_pos = start_pos + segment_len
    # make a copy so changes to 'segments' doesn't modify the original ekg_data
    segment = np.copy(bp[start_pos:end_pos])
    # if we're at the end and we've got a truncated segment, drop it
    if len(segment) != segment_len:
        continue
    segments.append(segment)

print("Produced %d waveform segments" % len(segments))
```
